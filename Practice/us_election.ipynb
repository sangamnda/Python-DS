```python
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('governors_county.csv')
df
```

```python
fig = px.bar(df, x = 'county', y = 'state',
             title='county state wise',
             labels={'state':'State', 'county':'County'},
             color = 'state')
fig.show()
```

```python
fig = px.pie(df, values= 'state', names= 'percent',
             title='Total Vote Percent by State',
             labels={'state':'State', 'percent':'Percent'},
             color= 'percent')
fig.show()
```

```python
fig = px.sunburst(df,
                  path = ['state', 'county'],
                  values = 'percent',
                  title = 'Total Votes by percent',
                labels = {'state': 'State', 'percent':'Percent'},
                color = 'percent',
                color_continuous_scale = px.colors.sequential.RdBu,
                height = 600, width = 800, template = 'plotly_dark')
fig.show()
```

```python
fig = px.sunburst(df,
                  path = ['state', 'county'],
                  values = 'total_votes',
                  title = 'Total Votes by State',
                labels = {'state': 'State', 'total_votes':'Total Votes'},
                color = 'total_votes',
                color_continuous_scale = px.colors.sequential.RdBu,
                height = 600, width = 800, template = 'plotly_dark')
fig.show()
```
