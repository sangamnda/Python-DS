```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
# histogram of age distribution
fig = sns.histplot(df, x = 'age', color = 'red', edgecolor = 'black')
fig.set_title('Age Distribution of Titatnic Passengers')
fig.set_xlabel('Age')
fig.set_ylabel('Frequency')
```

```python
# histogram of age distribution
fig = sns.histplot(df, x = 'age', color = 'red', edgecolor = 'black', kde = True)
fig.set_title('Age Distribution of Titatnic Passengers')
fig.set_xlabel('Age')
fig.set_ylabel('Frequency')
```

```python
# histogram of age distribution
fig = sns.histplot(df, x = 'age', color = 'red', edgecolor = 'black', kde = True, bins = 30)
fig.set_title('Age Distribution of Titatnic Passengers')
fig.set_xlabel('Age')
fig.set_ylabel('Frequency')
```

```python
# count of survival based on gender
fig = sns.countplot(df, x = 'survived', color = 'purple', edgecolor = 'black', hue = 'sex')
fig.set_title('Count of Titanic Passengers by Survival Status')
fig.set_xlabel('Survival Status')
fig.set_ylabel('Count')
```

```python
fig = sns.barplot(df, x = 'class', y = 'fare', hue = 'sex', palette='coolwarm')
fig.set_title('Avarage fare by class and gender')
fig.set_xlabel('class')
fig.set_ylabel('Avarage fare')
```

```python
sns.countplot(df, x = 'class', palette='viridis')
fig.set_title('count of passenfers by class')
fig.set_xlabel('class')
fig.set_ylabel('count')
```

```python
sns.pairplot(df[['age', 'fare', 'survived', 'pclass']], hue = 'survived', palette='terrain')
```
