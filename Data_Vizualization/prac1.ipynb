```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# creating a figure and axis
fig, ax = plt.subplots(figsize=(12,6))
# setting the background color
fig.set_facecolor('lightgray')
ax.set_facecolor('lightyellow')
# plotting the data
a3 = np.array([[[1,2],[3,4],[5,6],[7,8]]])
ax.plot(a3,
        marker = 'o',
        markersize = 7,
        markerfacecolor = 'red',
        markeredgecolor = 'blue',
        markeredgewidth = 1,
        linestyle = '-',
        linewidth = 1,
        color = 'Purple')
# setting the title and label
ax.set_title('LINE PLOT', fontsize = 20, color = 'red')
ax.set_xlabel('X-AXIS', fontsize = 15, color = 'blue')
ax.set_ylabel('Y-AXIS', fontsize = 15, color = 'blue')
```
