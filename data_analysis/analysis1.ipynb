```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('titanic')
df
```

```python
df.isnull().sum()
```

```python
avg_age = df['age'].mean()
print(avg_age)
```

```python
avg_age = round(df['age'].mean())
print(avg_age)
```

```python
df['age'].fillna(avg_age, inplace=True)
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.dropna(subset = ['embarked'], inplace = True)
```

```python
df.isnull().sum()
```

```python
deldeck = ['deck']
df = df.drop(columns=deldeck)
df.columns
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['alive'],inplace=True)
df
```
